```python
import numpy as np

import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
x = np.linspace(0, 10, 50)
y = 2 * x + 1 + np.random.normal(0, 3, size=x.shape)

# Create scatter plot
plt.scatter(x, y)
plt.title("Scatter Plot with Linear Relationship and Noise")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
